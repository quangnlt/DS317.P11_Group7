```python
def visualize_and_correlate_old(df):
    # Loại bỏ các cột 'id' và 'mssv' nếu chúng tồn tại
    df_clean = df.drop(['id', 'mssv'], axis=1, errors='ignore')

    # Loại bỏ các hàng có giá trị null
    df_clean = df_clean.dropna()

    # Boxplot cho các biến số
    numeric_cols = df_clean.select_dtypes(include=['float64', 'int64']).columns
    for col in numeric_cols:
        plt.figure(figsize=(10, 5))
        sns.boxplot(x=df_clean[col])
        plt.title(f'Boxplot of {col}')
        plt.show()

    # Bảng và trực quan hóa hệ số tương quan
    correlation_matrix = df_clean[numeric_cols].corr()
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
    plt.title('Correlation Matrix Heatmap')
    plt.show()

    # Barplot cho các biến phân loại
    categorical_cols = df_clean.select_dtypes(include=['object']).columns
    for col in categorical_cols:
        plt.figure(figsize=(10, 5))
        sns.countplot(x=df_clean[col], order=df_clean[col].value_counts().index)
        plt.title(f'Barplot of {col}')
        plt.xticks(rotation=45)
        plt.show()
```

```python
def visualize_and_correlate(df):
    # Loại bỏ các cột không cần thiết
    df_clean = df.drop(['id', 'mssv'], axis=1, errors='ignore')

    # Loại bỏ các bản ghi có giá trị thiếu
    df_clean = df_clean.dropna()

    # Thiết lập giao diện đẹp cho biểu đồ
    sns.set(style='whitegrid')

    # Trực quan hóa các thuộc tính số
    numeric_cols = df_clean.select_dtypes(include=['float64', 'int64']).columns
    for col in numeric_cols:
        distinct_values = df_clean[col].nunique()
        if distinct_values < 7:
            # Nếu có ít hơn 7 giá trị, sử dụng barplot nằm ngang
            plt.figure(figsize=(10, 5))
            sns.countplot(y=df_clean[col], order=df_clean[col].value_counts().index)
            plt.title(f'Biểu đồ tần suất của cột {col}')
            plt.xlabel('Số lượng')  # Đổi tên trục x thành 'Số lượng'
            plt.show()
        else:
            # Nếu nhiều giá trị hơn, sử dụng biểu đồ hộp
            plt.figure(figsize=(10, 5))
            sns.boxplot(x=df_clean[col])
            plt.title(f'Biểu đồ phân phối theo dạng hộp của cột {col}')
            plt.show()

    # Tạo và trực quan hóa ma trận hệ số tương quan
    correlation_matrix = df_clean[numeric_cols].corr()
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
    plt.title('Ma trận hệ số tương quan')
    plt.show()

    # Trực quan hóa các thuộc tính phân loại
    categorical_cols = df_clean.select_dtypes(include=['object']).columns
    for col in categorical_cols:
        distinct_values = df_clean[col].nunique()
        if distinct_values > 20:
            # Hiển thị top 20 giá trị phổ biến nhất bằng barplot nằm ngang
            top_20 = df_clean[col].value_counts().nlargest(20)
            plt.figure(figsize=(10, 5))
            sns.barplot(y=top_20.index, x=top_20.values, orient='h')
            plt.title(f'Biểu đồ tần suất của cột {col} (Top 20)')
            plt.xlabel('Số lượng')  # Đổi tên trục x thành 'Số lượng'
            plt.show()
        else:
            # Hiển thị tất cả các giá trị bằng barplot nằm ngang
            plt.figure(figsize=(10, 5))
            sns.countplot(y=df_clean[col], order=df_clean[col].value_counts().index)
            plt.title(f'Biểu đồ tần suất của cột {col}')
            plt.xlabel('Số lượng')  # Đổi tên trục x thành 'Số lượng'
            plt.show()
```

# Phân tích sơ bộ về các yếu tố chất lượng dữ liệu:
## Độ chính xác
- Tỷ lệ chính xác: Chưa thể xác định rõ do thiếu thông tin so sánh với dữ liệu thực tế, nhưng ngoại lệ và giá trị âm trong các điểm số (như điểm rèn luyện) cần được xem xét thêm để đảm bảo tính chính xác. Ngoài ra cần kiểm tra lại đảm bảo các giá trị đúng với cột tương ứng của chúng (ví dụ như cột ghichu bảng uitdrl)
- Độ lệch chuẩn: Các điểm số (qt, gk, ck) và tín chỉ có sự phân tán lớn, cần xem xét thêm để đánh giá độ chính xác.
## Tính đầy đủ
- Dữ liệu thiếu: Một số biến như tình trạng và trạng thái có nhiều giá trị đơn lẻ, cho thấy khả năng thiếu dữ liệu.
- Đồng bộ giữa các bảng: Các bảng liên quan đến điểm và tín chỉ cần được đối chiếu để đảm bảo tất cả sinh viên đều có thông tin đầy đủ (mã sinh viên, điểm, khóa học, tín chỉ).
## Tính nhất quán
- Mâu thuẫn dữ liệu: Các giá trị ngoại lệ trong điểm số và tín chỉ (như sotchk trên 150) cần được kiểm tra tính hợp lệ.
## Tính kịp thời
- Thời điểm thu thập dữ liệu: Phần lớn dữ liệu thuộc các năm từ 2014-2017, nhưng chưa rõ dữ liệu có được cập nhật gần đây không.
- Tần suất cập nhật: Không có thông tin về lần cập nhật cuối cùng, cần kiểm tra tính kịp thời.
## Tính hợp lệ
- Danh sách giá trị hợp lệ: Các trạng thái và tình trạng cần được đối chiếu với danh sách giá trị hợp lệ (nếu có).
## Tính duy nhất
- Mã sinh viên: Cần kiểm tra tính duy nhất của mã sinh viên (mssv) để đảm bảo không có trùng lặp.
- Các giá trị trùng lặp: Số tín chỉ học kỳ và điểm tổng tích lũy có nhiều giá trị bất thường, cần đối chiếu để loại bỏ bản ghi trùng hoặc sai sót.
