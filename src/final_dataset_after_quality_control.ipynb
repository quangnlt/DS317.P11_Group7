# Tạo ra cột điểm Trung Bình cho 8 kỳ mà sinh viên học

## Tính học kỳ liên tục

Giả sử  bắt đầu học năm 2021 và hiện tại là học kỳ 1 năm 2024, tức là sinh viên này  đã học được 7 học kỳ (học kỳ 1 và học kỳ 2 của các năm 2021, 2022, 2023 và học kỳ 1 năm 2024).

### Xác định năm học đầu tiên (namhoc_min) của mỗi sinh viên

```python
import pandas as pd
```

```python
df_final = pd.read_csv('df_final.csv')
```

```python
df_final['namhoc_min'] = df_final.groupby('mssv')['namhoc'].transform('min')
```

### Tính học kỳ gom (hocky_gom)

```python
df_final['hocky_gom'] = (df_final['namhoc'] - df_final['namhoc_min']) * 2 + df_final['hocky']
```

```python
# Kiểm tra kết quả
print(df_final[['mssv', 'namhoc', 'hocky', 'namhoc_min', 'hocky_gom']].head())
```

## Pivot dữ liệu để tạo các cột diemtbhk_1, diemtbhk_2, ...

```python
pivot_df = df_final.pivot_table(
    values='dtbhk',  # Giá trị cần pivot (điểm trung bình học kỳ)
    index='mssv',    # Mỗi sinh viên là 1 dòng
    columns='hocky_gom',  # Các cột là học kỳ gom
    aggfunc='first'  # Lấy giá trị đầu tiên nếu có nhiều bản ghi
)
```

```python
# Đổi tên cột thành diemtbhk_1, diemtbhk_2, ...
pivot_df.columns = [f'diemtbhk_{int(col)}' for col in pivot_df.columns]
```

## Merge với DataFrame chính để kết hợp dữ liệu

```python
final_df = pd.merge(df_final[['mssv', 'namsinh', 'gioitinh', 'khoa', 'hedt','drl','diem_tt' ,'khoahoc', 'dtb_toankhoa', 'dtb_tichluy', 'sotc_tichluy']].drop_duplicates('mssv'),
                     pivot_df,
                     on='mssv', how='left')
```

```python
# Kiểm tra dữ liệu kết quả
final_df.head(50)
```

## Xóa đi các cột điểm trung bình học kỳ 9 tới 19

```python
# Create a list of column names to be dropped
cols_to_drop = [f'diemtbhk_{i}' for i in range(9, 20)]

# Drop the specified columns from the DataFrame
final_df = final_df.drop(columns=cols_to_drop)
```

# Kiểm soát chất lượng dữ liệu

## Điền NaN namsinh = mean

```python
mean_nam_sinh = int(final_df['namsinh'].mean(skipna=True))
```

```python
final_df['namsinh'] = final_df['namsinh'].fillna(mean_nam_sinh)
```

```python
final_df['namsinh'] = final_df['namsinh'] - final_df['namsinh'].min()
```

## Bỏ bớt một số thuộc tính

```python
final_df = final_df.drop(['khoa', 'hedt', 'khoahoc'], axis=1)
```

## Điền NaN gioitinh = "Nam"

```python
final_df['gioitinh'] = final_df['gioitinh'].fillna('Nam')
```

```python
final_df['gioitinh'] = final_df['gioitinh'].replace({'Nam': 0, 'Nữ': 1})
```

## Xử lý NaN diem_tt

```python
# Điền NaN diem_tt bằng 30
final_df['diem_tt'] = final_df['diem_tt'].fillna(30)
```

```python
final_df.info()
```

## Xử lý NaN DRL

```python
# Xử lý NaN cột drl bằng mean
mean_drl = final_df['drl'].mean(skipna=True)
final_df['drl'] = final_df['drl'].fillna(mean_drl)
```

## Xử lý NaN các cột dtbhk

```python
# Danh sách các cột `diemtbhk`
diemtbhk_cols = [col for col in final_df.columns if col.startswith('diemtbhk_')]
```

```python
# Hàm điền giá trị NaN bằng trung bình của các kỳ có dữ liệu
def fill_nan_with_overall_avg(row):
    # Tính trung bình các kỳ đã có điểm
    avg_score = row[diemtbhk_cols].dropna().mean()
    # Điền giá trị NaN bằng trung bình
    row[diemtbhk_cols] = row[diemtbhk_cols].fillna(avg_score)
    return row
```

```python
# Áp dụng cho từng dòng dữ liệu (theo từng sinh viên)
final_df = final_df.apply(fill_nan_with_overall_avg, axis=1)
```

## Xử lý NaN cho dtb_toankhoa và dtb_tichluy bằng cách tính trung bình cộng các điểm ở mỗi kỳ của mỗi sv

```python
# Bước 1: Tính điểm trung bình toàn khóa (dtb_toankhoa) cho từng sinh viên
dtb_toankhoa = final_df.groupby('mssv')['dtb_toankhoa'].mean().reset_index(name='dtb_toankhoa')
```

```python
# Bước 2: Tạo cột mới dtb_tichluy trong DataFrame chính
# Dùng hàm map để gán giá trị dtb_toankhoa tương ứng cho từng sinh viên
final_df['dtb_toankhoa'] = final_df['mssv'].map(dtb_toankhoa.set_index('mssv')['dtb_toankhoa'])
final_df['dtb_tichluy'] = final_df['dtb_toankhoa']  # Cột tích lũy tương tự dtb_toankhoa
```

```python
# Bước 3: Hiển thị kết quả
print("DataFrame sau khi thêm cột dtb_toankhoa và dtb_tichluy:")
print(final_df)
```

## Xóa các hàng không có giá trị

```python
# Xóa các hàng có toàn NaN trong cột 'dtbhk' của tất cả các cột diemtbhk_i
diemtbhk_cols = [col for col in final_df.columns if col.startswith('diemtbhk_')]
final_df.dropna(subset=diemtbhk_cols, how='all', inplace=True)
```
